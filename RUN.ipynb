```python
num_class = 10
# 9, 18, 27, 36
num_nodes_per_class = 9

dataset_name = 'mnist'
#dataset_name = 'fashion_mnist'

l_ideal = []
for p1 in range(num_class):
    l2 = []
    for p2 in range(num_nodes_per_class):  
        l2.append(str(p1)+"_"+str(p2))
    l_ideal.append(l2)
l_ideal
```

```python
filename = 'clust_'+dataset_name+str(num_nodes_per_class)+'.txt'
with open(filename) as file:
    lines = file.readlines()[1:]
    l_clust = [sorted(line.strip().replace(' ', '').split(',')) for line in lines]

for i in l_clust:
    print(i)
```

```python
%%capture
liss = l_clust; out_fold = 'clust'
for num,i in enumerate(liss[0:]):
    print(str(i))
    ll = ['./run_new.sh', dataset_name, str(5000+num), str(num_nodes_per_class), out_fold] + i
    as_string = " ".join(ll)
    subprocess.check_call(as_string, shell=True)
    #subprocess.Popen(as_string, shell=True)
    #time.sleep(5)

liss = l_ideal; out_fold = 'ideal'
for num,i in enumerate(liss[0:]):
    print(str(i))
    ll = ['./run_new.sh', dataset_name, str(5000+num), str(num_nodes_per_class), out_fold] + i
    as_string = " ".join(ll)
    subprocess.check_call(as_string, shell=True)
    #subprocess.Popen(as_string, shell=True)
    #time.sleep(5)

```

```python
#dataset_name = 'mnist'
#dataset_name = 'fashion_mnist'
#num_nodes_per_class = 18


pth = 'out'+str(num_nodes_per_class)+'/'
#pth = 'out9/'

import pandas as pd
import glob
files = glob.glob(pth+"output_local/"+dataset_name+"/*.csv")

df1 = pd.DataFrame()
for f in files:
    csv = pd.read_csv(f, index_col=[0])
    df1 = pd.concat([df1,csv])
    df1 = df1.sort_values('dataset')
df1 = df1.add_suffix('_local').reset_index(drop=True)
```

```python
import pandas as pd
import glob
files = glob.glob(pth+"output_clust/"+dataset_name+"/*.csv")

df2 = pd.DataFrame()
for f in files:
    csv = pd.read_csv(f,index_col=[0])
    df2 = pd.concat([df2,csv])
    df2 = df2.sort_values('dataset')
df2 = df2.add_suffix('_clust').reset_index(drop=True)
df2
```

```python
import pandas as pd
import glob
files = glob.glob(pth+"output_ideal/"+dataset_name+"/*.csv")

df3 = pd.DataFrame()
for f in files:
    csv = pd.read_csv(f,index_col=[0])
    df3 = pd.concat([df3,csv])
    df3 = df3.sort_values('dataset').reset_index(drop=True)
df3 = df3.add_suffix('_ideal').reset_index(drop=True)
```

```python
pd.concat([df1,df2,df3],axis=1)
```
